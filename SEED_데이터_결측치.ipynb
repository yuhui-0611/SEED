# 7_1

CBAM 시행 이전 식(년도 : 2001~2021)

```python
df_7_1 = pd.read_csv('/content/drive/MyDrive/SEED/equation_7_1.csv')
```

```python
df_7_1.columns
```

```python
inter_7_1 = df_7_1[['ln_TARIFF_ij*ln_FIT_i',
                    'ln_TARIFF_ij*ln_PATENTS_i',
                    'ln_TARIFF_ij*ln_RENEWABLE_i']]
```

```python
df_7_1_except_inter = df_7_1[list(set(df_7_1.columns) - set(inter_7_1.columns))]
```

```python
df_7_1_except_inter.head()
```

국가쌍으로 봤을 때의 결측

```python
isna_df_7_1 = df_7_1_except_inter.isna()
isna_df_7_1.head()
```

```python
missing_7_1 = isna_df_7_1.groupby([df_7_1_except_inter['i'], df_7_1_except_inter['j']]).sum()
missing_7_1
```

행 더하기

```python
pair_total_7_1 = (missing_7_1.sum(axis=1).rename('total_missing').reset_index())
pair_total_7_1
```

한눈에 보기 쉽게 matrix로 바꾸기

## 국가 matrix

```python
pair_matrix_7_1 = pair_total_7_1.pivot(index='j', columns='i', values='total_missing')
pair_matrix_7_1
```

**위 matrix에서 결측값에 규칙성이 보이는 국가(열)만 일단 추출**
